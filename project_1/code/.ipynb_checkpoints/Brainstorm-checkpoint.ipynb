# Problem Statement 1: Number of rainy days and intense rainfall can affect economy

Data needed to evaluate economy impact: Sales? GDP?

https://eos.org/articles/rainy-days-dampen-economic-growth

High-income countries are the hardest hit by daily rainstorms because the bedrock industries of their economies—manufacturing and service

Previous work by Kotz found that temperature variability from one day to the next reduces macroeconomic growth.

    “Days with any considerable amount of rainfall constitute suboptimal economic conditions.”

The latest work found a similar connection between daily precipitation patterns and economic production. An increase in days with more than 1 millimeter of rain—a relatively small amount—showed a strong reduction in the rate of economic growth. “This suggests that days with any considerable amount of rainfall constitute suboptimal economic conditions,” wrote the authors in Nature. The relationship held even for days with just 0.1 millimeter of rain.

Days with very high amounts of rain—exceeding the 99.9th percentile of data spanning 40 years—further dampen growth.

Extreme rain disrupts daily life by washing out roads, flooding buildings, and causing power outages.

```python
import requests # HTTP requests (GET / POST)
import datetime
import pandas as pd
from IPython.display import display, Image
import folium
import json
```

```python
# Obtaining today's PSI
today = datetime.datetime.today()
params = {"date": today.strftime("%Y-%m-%d")} # YYYY-MM-DD 
psi = requests.get('https://api.data.gov.sg/v1/environment/psi', params=params).json()
```

```python
# Showing latest PSI in a dataframe
print(psi['items'][0]['timestamp']) # 12am GMT+8
pd.DataFrame(psi['items'][0]['readings']) # 
```

```python
# Obtaining temperature data
today = datetime.datetime.today()
params = {"date_time": today.strftime("%Y-%m-%dT%H:%M:%S")} # YYYY-MM-DD 
wx = requests.get('https://api.data.gov.sg/v1/environment/air-temperature', params=params).json()
```

```python
tempDf = pd.DataFrame(wx['metadata']['stations']).merge(pd.DataFrame(wx['items'][0]['readings']), left_on='id', right_on='station_id', how='inner')
display(tempDf)
tempDf = tempDf[["station_id", "name", "value", "location"]]
tempDf.columns = ['Station ID', 'Name', 'Temperature', "Location"]
tempDf
```

```python
# Taxi availability data
today = datetime.datetime.today()
print(today)
print(today.strftime("%Y-%m-%d"))
params = {"date": today.strftime("%Y-%m-%d")} # YYYY-MM-DD
taxi = requests.get('https://api.data.gov.sg/v1/transport/taxi-availability', params=params).json()
taxi['features'][0]['properties']['taxi_count']
```

```python
# Plotting first 100 taxis only
sg_map = folium.Map([1.3521, 103.8198], zoom_start = 11, tiles="Stamen Terrain")
print(taxi['features'][0]['geometry']['coordinates'])
for coord in taxi['features'][0]['geometry']['coordinates'][:1000]:
    print([coord[1], coord[0]])
    folium.Marker([coord[1], coord[0]]).add_to(sg_map)
sg_map
```
